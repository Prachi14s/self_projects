## calculatimg correlation of skills w.r.t. thier project team no. and team members.
## Finding a rank for all teams

```python
final_df.set_index('TEAM MEMBERS').stack().groupby(level=0).value_counts().unstack(fill_value=0)
```

```python
final_df.set_index('DESIGNATION').stack().groupby(level=0).value_counts().unstack(fill_value=0)
```

### It is showing correlation of skills w.r.t. designation, project team no., team member which will lead to a rank to every team with maximum true and minimum false

```python
final_teams_rank=final_df.set_index('PROJECT TEAM NO.').stack().groupby(level=0).value_counts().unstack(fill_value=0)
final_teams_rank
```

```python
final_teams_rank['score_ranked']=final_teams_rank['True'].rank(ascending=0,method='dense')
final_teams_rank
```

## Here it only teams 1, 6, 7, 8 are showing rank 1 because they are matching skill set, and domain.
## Other teams are not considerable due to different domains, skills not matched. So they all are showing at same rank.
